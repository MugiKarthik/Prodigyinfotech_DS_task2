# EDA

# Univarint

continuous variable

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Define attribute names and their corresponding min/max values
attributes = {
    'Age': (0, 80),
    'SibSp': (0, 8),
    'Parch': (0, 9),
    'Fare': (0, 520)
}

# Calculate the number of rows and columns for subplots
num_attributes = len(attributes)
num_cols = 2  # Number of columns for subplots
num_rows = (num_attributes + num_cols - 1) // num_cols

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))

# Plot distribution for each attribute
for i, (attribute, (min_value, max_value)) in enumerate(attributes.items()):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    
    # Plot histogram
    n, bins, patches = ax.hist(data[attribute], bins=10, range=(min_value, max_value), edgecolor='black', alpha=0.75)
    ax.set_xlabel(attribute)
    ax.set_ylabel('Percentage')
    ax.set_title(f'Distribution of {attribute}')
    ax.grid(axis='y', alpha=0.75)
    
    # Add percentage labels to the chart
    total = sum(n)
    for j, patch in enumerate(patches):
        percentage = '{:.1f}%'.format(100 * n[j] / total)
        x = patch.get_x() + patch.get_width() / 2
        y = patch.get_height()
        ax.text(x, y, percentage, ha='center', va='bottom')
    
    # Set y-axis ticks as percentages
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:.0%}'.format(y/total)))
    ax.set_ylim(0, max(n)*1.2)

# Adjust layout
plt.tight_layout()
plt.show()
```

# Catergorical variables

```python
import matplotlib.pyplot as plt
import numpy as np

# Define categorical attributes
categorical_attributes = ['Survived', 'Pclass', 'Sex', 'Embarked']

# Calculate the number of rows and columns for subplots
num_attributes = len(categorical_attributes)
num_cols = 3  # Number of columns for subplots
num_rows = (num_attributes + num_cols - 1) // num_cols

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 12))

# Plot pie chart for each categorical attribute
for i, attribute in enumerate(categorical_attributes):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    
    # Count the number of occurrences of each category
    counts = data[attribute].value_counts()
    
    # Create pie chart
    ax.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Distribution of {attribute}')
    
    # Equal aspect ratio ensures that pie is drawn as a circle
    ax.axis('equal')

# Adjust layout
plt.tight_layout()
plt.show()
```

# bivariant analysis
continuous variables

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define continuous attributes and their ranges
continuous_attributes = {
    'Age': (0, 80),
    'SibSp': (0, 8),
    'Parch': (0, 9),
    'Fare': (0, 520)
}

# Calculate quartile values for each continuous attribute grouped by 'Loan_Status'
quartiles = data.groupby('Survived').describe()

# Create subplots
fig, axes = plt.subplots(nrows=len(continuous_attributes), ncols=1, figsize=(10, 20))

# Plot boxplots for each continuous attribute
for i, (attribute, (min_value, max_value)) in enumerate(continuous_attributes.items()):
    # Create boxplot
    sns.boxplot(data=data, x='Survived', y=attribute, ax=axes[i])
    
    # Add quartile values to the plot
    for j, label in enumerate(quartiles.index):
        q1 = quartiles.loc[label, (attribute, '25%')]
        median = quartiles.loc[label, (attribute, '50%')]
        q3 = quartiles.loc[label, (attribute, '75%')]
        axes[i].text(j, q1, f'Q1: {q1:.2f}', ha='center', va='bottom', color='black')
        axes[i].text(j, median, f'Median: {median:.2f}', ha='center', va='bottom', color='black')
        axes[i].text(j, q3, f'Q3: {q3:.2f}', ha='center', va='bottom', color='black')

    # Set the title and axes labels
    axes[i].set_title(f'{attribute} Distribution by Survived Status')
    axes[i].set_xlabel('Survived')
    axes[i].set_ylabel(attribute)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define categorical attributes
categorical_attributes = [ 'Pclass', 'Sex', 'Embarked']

# Create subplots
fig, axes = plt.subplots(nrows=len(categorical_attributes), ncols=1, figsize=(8, 20))

# Calculate the count and ratio for each attribute and Loan_Status combination
for i, attribute in enumerate(categorical_attributes):
    percentData = data.groupby([attribute, 'Survived']).size().reset_index(name='n')
    percentData['ratio'] = percentData.groupby(attribute)['n'].apply(lambda x: 100 * x / x.sum())

    # Create the stacked bar chart with percentage labels
    bar_plot = percentData.pivot(index=attribute, columns='Survived', values='ratio').plot.bar(
        stacked=True, ax=axes[i]
    )

    for rect in axes[i].containers:
        axes[i].bar_label(rect, label_type='edge', labels=np.round(rect.datavalues, 1), padding=3)

    # Set the axis titles and legend
    axes[i].set_title(f'{attribute} Distribution by Survived')
    axes[i].set_xlabel(attribute)
    axes[i].set_ylabel('Percentage')
    handles, _ = axes[i].get_legend_handles_labels()
    labels = ['0 = Not Approved', '1 = Approved']
    axes[i].legend(handles, labels, title='Survived Status', bbox_to_anchor=(1, 1), loc='upper left')

# Adjust layout
plt.tight_layout()
plt.show()
```

# logistic regression

```python
data.replace({'Sex':{'male':1,'female':0},'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
data.head()
```

```python
#model#
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc,recall_score,precision_score


X = data.drop("Survived", axis=1)
y = data["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc,recall_score,precision_score
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import roc_curve, auc

# Fit the logistic regression model on the training data
model = sm.Logit(y_train, X_train)
result = model.fit()
result.summary()
```

```python
print("AIC:", result.aic)
# Add odds ratio to the output
conf = result.conf_int()
conf['Odds Ratio'] = np.exp(result.params)
conf.columns = ['Lower CI', 'Upper CI', 'Odds Ratio']
print(np.exp(result.params))
print(conf)
```

```python
# Make predictions on the test data
y_prob = result.predict(X_test)

# Calculate the false positive rate, true positive rate, and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Find the optimal cut value that maximizes the Youden's index (i.e., sensitivity + specificity - 1)
youden_index = tpr - fpr
optimal_idx = np.argmax(youden_index)
optimal_threshold = thresholds[optimal_idx]

# Make binary predictions based on the cut value
y_pred = (y_prob >= 0.5).astype(int)


# Calculate the area under the ROC curve
roc_auc = auc(fpr, tpr)


# plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--', label="Random guess")
plt.plot([0, fpr[optimal_idx]], [tpr[optimal_idx], tpr[optimal_idx]], 'r--', label="Optimal cut value")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("Receiver operating characteristic (ROC) curve")
plt.legend()
plt.show()
```

```python
from tabulate import tabulate
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
ppv = tp / (tp + fp)
npv = tn / (tn + fn)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)


table = [["Metric", "Value"],
         ["Sensitivity", sensitivity],
         ["Specificity", specificity],
         ["Positive Predictive Value", ppv],
         ["Negative Predictive Value", npv],
         ["Accuracy", accuracy],
         ["Precision", precision]]

print(tabulate(table, headers="firstrow"))
```

```python
# Evaluate the performance of the model
# plot the confusion matrix as a heatmap
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap="coolwarm", fmt="d")
plt.title("Confusion matrix")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```
